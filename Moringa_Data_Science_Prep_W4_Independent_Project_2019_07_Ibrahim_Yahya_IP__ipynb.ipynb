```
# Changing the case of column names to lower, stripping whitespaces, removing periods and replacing spaces with underscore
#
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_').str.replace('.', '')
```

```
df.columns
```

```
#Outliers
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)

IQR = q3 - q1

((df < q1 - 1.5 * IQR) | (df > q3 + 1.5 * IQR))
```

```
df = df[ ~ ((df < q1 - 1.5 * IQR) | (df > q3 + 1.5 * IQR)).any(axis = 1)]
```

```
#Identify the most popular hour of the day for picking up a shared electric car (Bluecar) 
#in the city of Paris over the month of April 2018


#df[(df.city == 'Paris') & (df.bluecar_counter < 4 )].groupby('hour')['address'].head(1)

df[(df.city == 'Paris') & (df.bluecar_counter < 7)].groupby('hour')['hour'].count().sort_values(ascending = False).head(1)
```

```
#What is the most popular hour for returning cars?
df[df.bluecar_counter != 7].groupby('hour')['hour'].count().sort_values(ascending = False).head(1)
```

```
#What station is the most popular?
#Overall?

df[(df.kind == 'STATION') & (df.status == 'ok')].groupby('public_name')[['kind']].count().sort_values(by = 'kind', ascending  = False).head(1)
```

```
# Identifying the station that is the most popular at the most popular picking hour
df[(df.kind == 'STATION') & (df.hour == 19)  & (df.status == 'ok')].groupby('public_name').count()[['kind']].sort_values(by = 'kind', ascending = False).head(1)
```

```
# Identifying the postal code which is the most popular for picking up Bluecars - overall


newdf = df.groupby(['hour' , 'postal_code' , 'public_name'])[['postal_code']].count().head(1)
```
